```python
# Install the OpenAI library if you haven't already
!pip install -q openai

import os
from openai import OpenAI

# --- Configuration for OpenAI API ---
# IMPORTANT: Replace 'YOUR_OPENAI_API_KEY' with your actual OpenAI API key.
# You can get one from https://platform.openai.com/account/api-keys
# It's recommended to store API keys securely, e.g., as environment variables or in Colab Secrets.
# For this example, we'll put it directly here, but be cautious in production.

# Option 1: Direct assignment (least secure, but simple for demonstration)
# os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_API_KEY"

# Option 2: Using Colab Secrets (recommended for Colab)
# from google.colab import userdata
# os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY') # Assuming you saved your key as 'OPENAI_API_KEY' in Colab Secrets

# Placeholder for API key. Please uncomment and use one of the options above.
client = OpenAI(api_key="im unable to generate my api key because i already have use its free version") # Replace with your actual key or use os.environ.get("OPENAI_API_KEY")

# --- Prompt Engineering: Define the chatbot's persona and safety guidelines ---
system_message = (
    "You are a helpful, empathetic, and knowledgeable medical assistant designed to provide "
    "general health information. Your primary goal is to educate users on common health topics "
    "and promote understanding, not to diagnose, treat, or give specific medical advice. "
    "Always preface your responses with a clear disclaimer that you are an AI and cannot "
    "provide medical advice, and that users should consult with a qualified healthcare professional "
    "for any health concerns."
    "When answering questions, focus on providing general, factual, and widely accepted health information. "
    "Avoid making recommendations for specific treatments, medications, or dietary changes. "
    "If a question seems to require personal medical advice or could be harmful if misinterpreted, "
    "gently redirect the user to consult a doctor or healthcare provider. "
    "Maintain a kind and professional tone. Do not generate code or external links."
)

# --- Function to interact with the LLM ---
def get_health_response(user_query):
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo", # Or "gpt-4" if you have access and prefer
            messages=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": user_query}
            ],
            temperature=0.7, # Controls randomness: lower for more deterministic output
            max_tokens=300 # Limit the length of the response
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}. Please ensure your API key is correct and you have network access."

# --- Simple Chatbot Loop ---
print("Hello! I'm your AI Medical Assistant. I can provide general health information.")
print("Remember, I cannot give medical advice. Please consult a healthcare professional for any specific health concerns.")
print("Type 'exit' to end the chat.")

while True:
    user_input = input("\nYour question: ")
    if user_input.lower() == 'exit':
        print("Thank you for chatting! Stay healthy.")
        break

    if not user_input.strip(): # Handle empty input
        print("Please ask a question.")
        continue

    # Add a safety pre-check for very sensitive keywords (basic, can be expanded)
    sensitive_keywords = ["diagnose me", "what should I take", "am I sick", "treat my", "cure my"]
    if any(keyword in user_input.lower() for keyword in sensitive_keywords):
        print("As an AI, I cannot diagnose conditions or recommend treatments. Please consult a qualified healthcare professional for personalized medical advice.")
        continue

    print("AI Assistant: Please remember, I am an AI and cannot provide medical advice. Consult a healthcare professional for any health concerns. " + get_health_response(user_input))
```
